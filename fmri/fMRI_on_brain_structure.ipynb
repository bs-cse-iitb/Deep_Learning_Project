```python
import nibabel as nib
from nilearn import plotting


fmri_file_4d = 'betas_session01.nii.gz'
img_4d = nib.load(fmri_file_4d)

# Specify a time point for visualization 
time_point = 100
img_at_time_point = nib.funcs.four_to_three(img_4d)[time_point]

# Plotting the data at the specified time point on a brain template
plotting.plot_stat_map(img_at_time_point, title=f'fMRI Activation Map at Time Point {time_point}', threshold=3, cut_coords=(0, 0, 0))


plotting.show()
```
